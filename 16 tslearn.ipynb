```python
# load and plot
from aeon.datasets import load_from_ts_file
import matplotlib.pyplot as plt

xc_tr, yc_tr = load_from_ts_file("./dataset/KDD_UTSC_TRAIN.ts", return_type="numpy2D")
xc_te, yc_te = load_from_ts_file("./dataset/KDD_UTSC_TEST.ts", return_type="numpy2D")

xr_tr, yr_tr = load_from_ts_file("./dataset/KDD_UTSER_TRAIN.ts", return_type="numpy2D")
xr_te, yr_te = load_from_ts_file("./dataset/KDD_UTSER_TEST.ts", return_type="numpy2D")

fig, ax = plt.subplots(1, 2)
ax[0].plot(xc_tr[0])
ax[1].plot(xr_tr[0])
plt.show()
```

```python
# tslearn distance fixed
from tslearn.neighbors import KNeighborsTimeSeriesClassifier, KNeighborsTimeSeriesRegressor
from tslearn.svm import TimeSeriesSVC, TimeSeriesSVR
from tslearn.shapelets import LearningShapelets

knn_c = KNeighborsTimeSeriesClassifier(n_neighbors=2, metric="dtw")
knn_c.fit(xc_tr, yc_tr)

svc_c = TimeSeriesSVC()
svc_c.fit(xc_tr, yc_tr)

shp_c = LearningShapelets(n_shapelets_per_size={3: 1}, max_iter=1)
shp_c.fit(xc_tr, yc_tr)

knn_r = KNeighborsTimeSeriesRegressor(n_neighbors=2, metric="dtw")
knn_r.fit(xr_tr, yr_tr)

svr_r = TimeSeriesSVR()
svr_r.fit(xr_tr, yr_tr)
```

```python
# tslearn distance fixed
from tslearn.neighbors import KNeighborsTimeSeriesClassifier, KNeighborsTimeSeriesRegressor
from tslearn.svm import TimeSeriesSVC, TimeSeriesSVR
from tslearn.shapelets import LearningShapelets

knn_c = KNeighborsTimeSeriesClassifier(n_neighbors=2, metric="dtw")
knn_c.fit(xc_tr, yc_tr)

svc_c = TimeSeriesSVC()
svc_c.fit(xc_tr, yc_tr)

shp_c = LearningShapelets(n_shapelets_per_size={3: 1}, max_iter=1)
shp_c.fit(xc_tr, yc_tr)

knn_r = KNeighborsTimeSeriesRegressor(n_neighbors=2, metric="dtw")
knn_r.fit(xr_tr, yr_tr)

svr_r = TimeSeriesSVR()
svr_r.fit(xr_tr, yr_tr)
```

```python
# aeon feature
from aeon.classification.feature_based import Catch22Classifier, FreshPRINCEClassifier, SummaryClassifier
from aeon.regression.feature_based import Catch22Regressor, FreshPRINCERegressor, SummaryRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

rf_c = RandomForestClassifier(n_estimators=3)
rf_r = RandomForestRegressor(n_estimators=3)

c22_c = Catch22Classifier(estimator=rf_c)
c22_c.fit(xc_tr, yc_tr)

fp_c = FreshPRINCEClassifier(n_estimators=3)
fp_c.fit(xc_tr, yc_tr)

sum_c = SummaryClassifier(estimator=rf_c)
sum_c.fit(xc_tr, yc_tr)

c22_r = Catch22Regressor(estimator=rf_r)
c22_r.fit(xr_tr, yr_tr)

fp_r = FreshPRINCERegressor(n_estimators=3)
fp_r.fit(xr_tr, yr_tr)

sum_r = SummaryRegressor(estimator=rf_r)
sum_r.fit(xr_tr, yr_tr)
```

```python
# aeon dl
from aeon.classification.deep_learning import TimeCNNClassifier, FCNClassifier, MLPClassifier
from aeon.regression.deep_learning import TimeCNNRegressor, FCNRegressor, MLPRegressor

cnn_c = TimeCNNClassifier(n_epochs=5)
cnn_c.fit(xc_tr, yc_tr)

fcn_c = FCNClassifier(n_epochs=5)
fcn_c.fit(xc_tr, yc_tr)

mlp_c = MLPClassifier(n_epochs=5)
mlp_c.fit(xc_tr, yc_tr)

cnn_r = TimeCNNRegressor(n_epochs=5)
cnn_r.fit(xr_tr, yr_tr)

fcn_r = FCNRegressor(n_epochs=5)
fcn_r.fit(xr_tr, yr_tr)

mlp_r = MLPRegressor(n_epochs=5)
mlp_r.fit(xr_tr, yr_tr)
```

```python
# aeon conv fix
from aeon.classification.convolution_based import RocketClassifier, HydraClassifier, MultiRocketHydraClassifier
from aeon.regression.convolution_based import RocketRegressor, MultiRocketHydraRegressor

roc_c = RocketClassifier(n_kernels=50)
roc_c.fit(xc_tr, yc_tr)

hyd_c = HydraClassifier()
hyd_c.fit(xc_tr, yc_tr)

mroc_c = MultiRocketHydraClassifier()
mroc_c.fit(xc_tr, yc_tr)

roc_r = RocketRegressor(n_kernels=50)
roc_r.fit(xr_tr, yr_tr)

mroc_r = MultiRocketHydraRegressor()
mroc_r.fit(xr_tr, yr_tr)
```

```python
# calc scores and plot
from sklearn.metrics import accuracy_score, mean_squared_error
import matplotlib.pyplot as plt

models_c = {
    "knn": knn_c, "svc": svc_c, "shapelet": shp_c,
    "rocket": roc_c, "hydra": hyd_c, "mrocket": mroc_c,
    "catch22": c22_c, "freshprince": fp_c, "summary": sum_c,
    "cnn": cnn_c, "fcn": fcn_c, "mlp": mlp_c
}

models_r = {
    "knn": knn_r, "svr": svr_r,
    "rocket": roc_r, "mrocket": mroc_r,
    "catch22": c22_r, "freshprince": fp_r, "summary": sum_r,
    "cnn": cnn_r, "fcn": fcn_r, "mlp": mlp_r
}

scores_c = {}
for name, model in models_c.items():
    try:
        scores_c[name] = accuracy_score(yc_te, model.predict(xc_te))
    except:
        scores_c[name] = 0

scores_r = {}
for name, model in models_r.items():
    try:
        scores_r[name] = mean_squared_error(yr_te, model.predict(xr_te))
    except:
        scores_r[name] = 0

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].barh(list(scores_c.keys()), list(scores_c.values()))
ax[0].set_title("accuracy (classification)")

ax[1].barh(list(scores_r.keys()), list(scores_r.values()))
ax[1].set_title("mse (regression)")

plt.tight_layout()
plt.show()
```
